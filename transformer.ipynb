```python
from transformers import pipeline
```

```python
import pandas as pd
```

```python
data=pd.read_html("https://en.wikipedia.org/wiki/National_Institutes_of_Technology_(India)",header=0)[2]
```

```python
print(data)
```

```python
data=data.astype(str)
```

```python
from transformers import AutoTokenizer,AutoModelForTableQuestionAnswering
```

```python
model='google/tapas-large-finetuned-wtq'
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
tqa= pipeline(task="table-question-answering",model=model,tokenizer=tokenizer)
```

```python
tqa(table=data,query='give the NIT nagpur abbrevation,founded,Establish and Overall rank')
```
